```python
import talib
```

```python
def load(df_norm_pct_change):
    df = pandas.DataFrame(index=df_norm_pct_change.index, columns=['closeOpen'])
    
    for i in range(1, len(df_norm_pct_change)):
        pct = ((df_norm_pct_change.ix[i]['open'] - df_norm_pct_change.ix[i - 1]['close']) / df_norm_pct_change.ix[i - 1]['close'])
        df['closeOpen'][i] = pct
    
    df['rolling_max'] = pandas.rolling_max(df['closeOpen'], window=5)
    df['rolling_min'] = pandas.rolling_min(df['closeOpen'], window=5)
    df['diff'] = talib.SMA(df['rolling_max'] - df['rolling_min'])

    return df
    

closeOpen = load(df_norm)
```
