File utils

```python
import os
import re


def _path_sort(path):
    return path.lower()


def listdir(s_dir, s_regex, f_sort=_path_sort):
    r_pattern = re.compile(s_regex, re.IGNORECASE)
    result = []
    
    for s_filename in os.listdir(s_dir):
        if r_pattern.search(s_filename):
            result.append(os.path.abspath(os.path.join(s_dir, s_filename)))

    return sorted(result, key=f_sort)
```

Data Structs

```python
class _LazyDict(object):

    def __init__(self, factory):
        self.c_dataMap = {}
        self.c_factory = factory

    def get(self, s_symbol):
        if self.c_dataMap.get(s_symbol) is None:
            self.c_dataMap[s_symbol] = self.c_factory.get()
        return self.c_dataMap[s_symbol]
    
    def keys(self):
        return self.c_dataMap.keys()


class _ListFactory(object):
    
    def get(self):
        return []
```

Cotahist

```python
import bovespaparser.bovespaparser as bp


def _importData():
    result = []
    for s_file in _get_cotahist_file_paths():
        with open(s_file, 'rU') as f:
            result.extend(bp.parsedata(f))
    return result


def _get_cotahist_file_paths():
    return listdir(_get_data_dir(), r'COTAHIST_A[\d]{4}\.txt$')
```

```python
import pandas

class CotahistImporter(object):

    def __init__(self, ll_rawdata):
        self.seriesMap = {}
        
        c_dataMap = _LazyDict(_ListFactory())
        mapping = [("open", 1), ("high", 2), ("low", 3), ("close", 4), ("volume", 5)]

        for s_symbol, date, f_open, f_min, f_max, f_close, i_vol in ll_rawdata:
            c_symbolData = c_dataMap.get(s_symbol)
            c_symbolData.append([date, f_open, f_max, f_min, f_close, i_vol])

        for s_symbol in c_dataMap.keys():
            ll_data = zip(*c_dataMap.get(s_symbol))
            ts_index = ll_data[0]
            self.seriesMap[s_symbol] = {s_name: pandas.TimeSeries(ll_data[i_columnIndex], index=ts_index, name=s_name) for s_name, i_columnIndex in mapping}

    def getTimeSeriesMap(self):
        return self.seriesMap
```

raw providers

```python
import pandas
import providers.base as base

class CotahistProvider(base.RawProvider):
    
    def __init__(self):
        self.d_ts = CotahistImporter(_importData()).getTimeSeriesMap()

    def load(self, s_symbol, da_start=None, da_end=None):
        df_symbol = pandas.DataFrame(self.d_ts[s_symbol], columns=['open', 'high', 'low', 'close'])
        if da_start and da_end:
            return df_symbol[da_start:da_end]
        if da_start:
            return df_symbol[da_start:]
        if da_end:
            return df_symbol[:da_end]
        return df_symbol
```
