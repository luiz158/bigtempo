```python
import random

class R(object):
    def __init__(self):
        self.last = 0

    def r(self):
        self.last = self.last - 50 + random.randint(0, 100)
        return self.last
```

```python
data = []

r = R()
for i in range(3000):
    data.append([i*(24*3600*1000), r.r(), r.r(), r.r(), r.r()])
```

```python
plot(options.Options('aleatorio').add(options.Axis('Axis1')).add(options.Series('Series1', data)))
```
