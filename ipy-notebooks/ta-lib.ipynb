```python
import numpy
import talib

data = numpy.random.random(100)
output = talib.SMA(data)
```

```python
import matplotlib
def plot(ll_data, s_symbol, s_xAxDesc, s_yAxDesc):
    pyplot.plot_date(ll_data[0], ll_data[1], '-', xdate=True)

    pyplot.title('%s - %s / %s' % (s_symbol, s_yAxDesc, s_xAxDesc))
    pyplot.xlabel(s_xAxDesc)
    pyplot.ylabel(s_yAxDesc)
    #pyplot.savefig('output/%s.png' % s_symbol)
    pyplot.show()
```

```python
from providers.raw import *
```

```python
data = YahooCotationProvider().load("PETR4")
data[3]
```

```python
#newd = pandas.TimeSeries([float(v) for v in data[3]], index=data[3].index, name=data[3].name+"new")
newd = pandas.TimeSeries(data[3], index=data[3].index, name=data[3].name+"new")
newd
```

```python
output = talib.SMA(newd)
output
```

```python
upper, middle, lower = talib.BBANDS(newd, matype=talib.MA_T3)
upperd = pandas.TimeSeries(upper, index=data[3].index, name="upper")
middled = pandas.TimeSeries(middle, index=data[3].index, name="middle")
lowerd = pandas.TimeSeries(lower, index=data[3].index, name="lower")
```

```python
upperd.plot()
```

```python
df = pandas.DataFrame({"close":newd, "upper":upperd, "middle":middled, "lower":lowerd}, data[3].index)
```

```python
df[0:100].plot()
```

```python
m1, m2, m3 = talib.MACD(newd)
sm1 = pandas.TimeSeries(m1, index=data[3].index, name="m1")
sm2 = pandas.TimeSeries(m2, index=data[3].index, name="m2")
sm3 = pandas.TimeSeries(m3, index=data[3].index, name="m3")
df = pandas.DataFrame({"macd":sm1, "signal":sm2, "histo":sm3}, data[3].index)
```

```python
df[0:100].plot()
```

```python
df['histo'][0:100].plot(kind='bar')
```
