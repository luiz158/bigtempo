```python
from datetime import date as date
import util.dateutils as dateutils
import providers.base as base
import providers.byproduct as byproduct
```

```python
class AbstractPoint(base.AbstractProvider):
    pass
```

```python
class MACD_Buy_Point(AbstractPoint):

    def __init__(self, locator):
        base.AbstractProvider.__init__(self)
        self.locator = locator
    
    def load(self, s_symbol, da_start=None, da_end=None):
        df_macd = self.locator.get(byproduct.MACDProvider).load(s_symbol, da_start, da_end).dropna()
        df_cota = self.locator.get(byproduct.NormalizedCotationProvider).load(s_symbol, da_start, da_end)
        
        last = 0
        for i in range(len(df_macd)):
            v = df_macd.ix[i]['macdhist']
            if last < 0 and v > 0:
                yield df_cota[df_cota.index == df_macd.ix[i].name]
            last = v

class MACD_Sell_Point(AbstractPoint):

    def __init__(self, locator):
        base.AbstractProvider.__init__(self)
        self.locator = locator
    
    def load(self, s_symbol, da_start=None, da_end=None):
        df_macd = self.locator.get(byproduct.MACDProvider).load(s_symbol, da_start, da_end).dropna()
        df_cota = self.locator.get(byproduct.NormalizedCotationProvider).load(s_symbol, da_start, da_end)
        
        last = 0
        for i in range(len(df_macd)):
            v = df_macd.ix[i]['macdhist']
            if last > 0 and v < 0:
                yield df_cota[df_cota.index == df_macd.ix[i].name]
            last = v
```

```python
class AbstractStrategy(base.AbstractProvider):
    pass
```

```python
class MACD_Buy_Sell_Strategy(AbstractStrategy):
    
    def load(self, s_symbol, da_start=None, da_end=None):
        buy = MACD_Buy_Point(c_locator)
        sell = MACD_Sell_Point(c_locator)
        
        result = []
        
        curr_bp = None
        sell_point_generator = sell.load(s_symbol, da_start, da_end)
        curr_sp = next(sell_point_generator)

        for buy_point in buy.load(s_symbol, da_start, da_end):
            
            if buy_point.ix[0].name < curr_sp.ix[0].name:
                continue

            curr_bp = buy_point
            
            for sell_point in sell_point_generator:
                if sell_point.ix[0].name > curr_bp.ix[0].name:
                    curr_sp = sell_point
                    result.append([curr_bp.ix[0].name, curr_sp.ix[0].name, curr_bp.ix[0]['close'], curr_sp.ix[0]['close'], (curr_sp.ix[0]['close'] - curr_bp.ix[0]['close']) / curr_bp.ix[0]['close']])
                    break;
            

        return result
```

```python
def return_in_period(l, p):
    result = []
    for i in range(len(l)-p):
        sl = l[i:i+p]
        result.append([sl, sum(sl)])
    return result
```
