```python
entrada = [8, 5, 6, 1, 3, 2, 7, 4, 9]


# big-O notation: O(n²)
def ordenar(lista):
    for j in range(0, len(lista) - 1):

        for i in range(0, len(lista) - 1):
            atual = lista[i]
            proximo = lista[i + 1]
            
            if atual > proximo:
                lista[i] = proximo
                lista[i + 1] = atual
            
            print 'Rodada %i: Passo %i:' % (j+1, i+1)
            print lista
    
    lista_ordenada = lista
    return lista_ordenada
```

```python
ordenar(entrada)
```
