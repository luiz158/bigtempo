```python
import sys
import pandas
import bigtempo.core as core
import bigtempo.tester as tester
```

```python
engine = core.DatasourceEngine()
```

```python
@engine.datasource('SAMPLE',
                   tags=['SAMPLE', 'DAILY'])
class Sample(object):

    def evaluate(self, context, symbol, start=None, end=None):
        return pandas.DataFrame.from_csv('SAMPLE{%s}.csv' % symbol)
```

```python
@engine.datasource('WEEKLY_SAMPLE',
                   dependencies=['SAMPLE'],
                   tags=['SAMPLE', 'WEEKLY'],
                   frequency='W')
class Weekly(object):

    def evaluate(self, context, symbol, start=None, end=None):
        return context.dependencies('SAMPLE').resample
```

```python
engine.get('SAMPLE').process('').plot()
```
